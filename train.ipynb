```python
data.drop(['Employee_ID','Name'],axis=1,inplace=True)
```

```python
data
```

```python
data['Gender']=data['Gender'].map({'Male':1,'Female':0})
```

```python
data['Education_Level']=data['Education_Level'].map({'Bachelor':1,'Master':2,'PhD':3})
```

```python
data=pd.get_dummies(data,columns=['Job_Title','Department','Location'],drop_first=True)
```

```python
data
```

```python
x=data.drop('Salary',axis=1)
y=data['Salary']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
model1=LinearRegression()
model1.fit(xtrain,ytrain)
predic1=model1.predict(xtest)
from sklearn.metrics import mean_absolute_error,r2_score
r2score=r2_score(ytest,predic1)
mae=mean_absolute_error(ytest,predic1)
print(r2score)
print(mae)
```

```python
print(pd.DataFrame({'Feature': xtrain.columns, 'Coefficient': model1.coef_}))
```

```python
import matplotlib.pyplot as plt

plt.scatter(ytest, predic1, alpha=0.5)
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary")
plt.show()

```

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

plt.figure(figsize=(10, 6))
sns.scatterplot(x=ytest, y=predic1, alpha=0.4, color='blue', label='Predicted Points')
sns.lineplot(x=ytest, y=ytest, color='red', label='Perfect Prediction (y = x)')  # Ideal line
plt.xlabel("Actual Salary")
plt.ylabel("Predicted Salary")
plt.title("Actual vs Predicted Salary with Ideal Line")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(ytest, kde=True, color='green', label='Actual Salary', bins=30)
sns.histplot(predic1, kde=True, color='orange', label='Predicted Salary', bins=30)
plt.xlabel("Salary")
plt.title("Distribution of Actual vs Predicted Salaries")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
